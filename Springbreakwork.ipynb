### read in elements and lola file


```python
# dictionary holding filename and according element name
names = {'LP_GRS_Th_Global_2ppd.tif': 'Thorium', 
         'LP_GRS_Fe_Global_2ppd.tif': 'Iron',
         'LP_GRS_H_Global_2ppd.tif' : 'Hydrogen',
         './temperature_xyz/resizedLOLA.xyz' : 'LOLA value' 
        }

# input file (tif or xyz) -> output pandas dataframe
def fileToDataframe(file):
    df = gr.from_file(file).to_pandas()
    df = df[["x", "y", "value"]].copy()
    df.columns = ["x", "y", names[file]]
    return df


# uses above function to create pandas dataframes (which is essential for all other functions)
if "LP_GRS_Th_Global_2ppd.tif" in names:
    df1 = fileToDataframe("LP_GRS_Th_Global_2ppd.tif")
if "LP_GRS_Fe_Global_2ppd.tif" in names:
    df2 = fileToDataframe("LP_GRS_Fe_Global_2ppd.tif")
if "LP_GRS_H_Global_2ppd.tif" in names:
    df3 = fileToDataframe("LP_GRS_H_Global_2ppd.tif")
if "./temperature_xyz/resizedLOLA.xyz" in names:
    df4 = fileToDataframe("./temperature_xyz/resizedLOLA.xyz")
```

#### 2D plotting function that takes (x, y).
#### e.g Hours vs Mean Temp plot

```python
hours = range(0,24)
temp_columns_names = completeDataframe.columns[2:]
min_temp = completeDataframe[temp_columns_names].min()
max_temp = completeDataframe[temp_columns_names].max()
avg_temp = completeDataframe[temp_columns_names].mean()
print(min_temp.head())
print(max_temp.head())
print(avg_temp.head())
```

```python
def plot_two_val(x,y,opacity):
    plt.scatter(x, y, c='blue', marker='x', alpha = opacity)
    plt.show()
```

##### (x,y) = (hour, temp) temp is either avg, min, or max temperature value of the entire region of interest.

```python
# min_temp vs hours
plot_two_val(hours, min_temp, 1)
```

```python
# avg_temp vs hours
plot_two_val(hours, avg_temp, 1)
```

```python
# max_temp vs hours
plot_two_val(hours, max_temp, 1)
```

#### 2D plotting function that takes (x,y,z). z value is used for color of the point at (x,y). This is like
#### displaying an image. But the data can be sparse.

```python
def plot_three_val(x, y, z, opacity):
    plt.scatter(x, y, c=z, marker='x', alpha = opacity)
    plt.show()
```

##### (x,y,z) = (lat, long, elevation). This should look almost same as DEM file.

```python
# x, y, lola value
plot_three_val(df4['x'], df4['y'], df4['LOLA value'], 1)
```

##### (x,y,z) = (lat, long, temp) temp is either 1) temperature at time t or 2) avg, min, or max temperature value for each (lat, long) over time.
